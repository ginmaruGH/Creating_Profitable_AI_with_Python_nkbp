### 4.2.1 不要な項目の削除

```python
# 余分な列の削除

# 「等室名」（「等室」と同じ）
df1 = df_titanic.drop('等室名', axis=1)

# 「乗船港」（「乗船港コード」と同じ）
df2 = df1.drop('乗船港', axis=1)

# 「生存可否」（「生存」と同じ）
df3 = df2.drop('生存可否', axis=1)

# 結果の確認
print(df3.shape)
display(df3.head())
```

### 4.2.2 欠損値の対応

```python
# 欠損値の確認
display(df3.isnull().sum())
```

```python
display(df3['デッキ'].value_counts())
```

方針の決定

- 乗船港コード: 
  - 件数が2件と少ない -> 行ごと削除する
- 年齢: 
  - 数値データであり、欠損行数が177件とかなり多い -> 他データの平均値で代用
- デッキ: 
  - コード値データであり、欠損行数が688行と相当多い -> 欠損を意味するダミーコードを振って全行処理対象とする

```python
# 乗船港コード: 件数が2件と少ない
# -> 行ごと削除する

# dropna関数を利用する
df4 = df3.dropna(subset=['乗船港コード'])
```

```python
# 年齢: 数値データであり欠損行数が177件とかなり多い
# -> 他データの平均値で代用

# 平均値の計算
age_average = df4['年齢'].mean()

# fillna関数の利用
df5 = df4.fillna({'年齢': age_average})
```

```python
# デッキ: コード値データであり欠損行数が688行と相当多い
# -> 欠損を意味するダミーコードを振って全行処理対象とする

# 書籍ではreplace関数を利用するとしていますが、動かなくなったので下記の実装に差し替えます。
# p.96 補足で説明している実装とほぼ同じです。
# (ダミーコードは 'N' とする)

df5['デッキ'] = df5['デッキ'].astype(object)
df6 = df5.fillna({'デッキ': 'N'})
```

```python
# 処理結果の確認
display(df6.isnull().sum())
display(df6.head())
display(df6.tail())
```

### 4.2.3 2値ラベルの数値化

- 性別: male / female
- 成人男子: True / False
- 独身: True / False

をそれぞれ 1/ 0 に数値化する

「性別」の数値化

```python
display(df6['性別'].value_counts())
```

```python
# 辞書mf_mapの定義
mf_map = {'male': 1, 'female': 0}

# map関数を利用して数値化
df7 = df6.copy()
df7['性別'] = df7['性別'].map(mf_map)

# 結果確認
display(df7.head())
```

「成人男子」、「独身」の数値化

```python
display(df7['成人男子'].value_counts())
display(df7['独身'].value_counts())
```

```python
# 辞書tf_mapの定義
tf_map = {True: 1, False: 0}

# map関数を利用して数値化
df8 = df7.copy()
df8['成人男子'] = df8['成人男子'].map(tf_map)

# map関数を利用して数値化
df9 = df8.copy()
df9['独身'] = df8['独身'].map(tf_map)

# 結果確認
display(df9.head())
```

### 4.2.4 多値ラベルの数値化

One-Hot エンコーディング

```python
# get_dummies関数のサンプル

w = pd.get_dummies(df9['男女子供'], prefix='男女子供')
display(w.head(10))
```

```python
# get_dummies関数でカテゴリ値をone hot vectorに展開する関数を定義
# df 対象データフレーム
# column 対象列

def enc(df, column):
    """One-Hotエンコーディング用の関数"""
    # One Hot Vector生成
    df_dummy = pd.get_dummies(df[column], prefix=column)
    # もと列の削除
    df_drop = df.drop([column], axis=1)
    # 削除したデータフレームと、One Hot生成列を連結
    df1 = pd.concat([df_drop, df_dummy], axis=1)
    return df1
```

男女子供

- man / woman / child

```python
# 項目値の確認
display(df9['男女子供'].value_counts())
```

```python
# One-Hotエンコーディング
df10 = enc(df9, '男女子供')

# 結果確認
display(df10.head())
```

乗船港コードとデッキ

```python
# One-Hot エンコード化

# 乗船港コード
df11 = enc(df10, '乗船港コード')

# デッキ
df12 = enc(df11, 'デッキ')

# 結果確認
display(df12.head())
```

### 4.2.5 正規化

normalization

- 入力変数$x$を、最小値が「0」、最大値が「1」になるように、1次関数を使って変換する
- $x = x_{max}$のとき、「1」
- $x = x_{min}$のとき、「0」
- 外れ値の影響を受けやすい
- 最大値と最小値がわかっている場合、使うことが多い

$$

\hat{x} = \dfrac{x - x_{min}}{x_{max} - x_{min}}
$$

standardization

- 入力変数$x$が、平均「0」、分散「1」の正規分布になるように、1次関数を使って変換する
- 外れ値を含んでいる場合、使うのが無難

$$
\hat{x} = \dfrac{x - m}{\sigma}
$$

```python
# standardization

from sklearn.preprocessing import StandardScaler

df13 = df12.copy()
stdsc = StandardScaler()
df13[['年齢', '料金']] = stdsc.fit_transform(df13[['年齢', '料金']])

# 結果確認
display(df13.head())
```
